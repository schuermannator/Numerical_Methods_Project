### Given:
- Insulated Rod around circumference
- x = 0 and x = L not insulated
- Rod is initially at Ti and in equilibrium with its surroundings

---

1. At time t = 0, surface temp of ends reaches 0 deg C
2. For t>0, temp distribution in the rod changes with time and approaches steady state
3. T(x,t) is to be determined numerically
4. Determine the internal energy of the rod as a function of time

---

### Properties:
- rho = 1000 kg/m^3
- k = 500 W/mK
- c = 500 J/kgK
- L = 1 m
- A = .001 m^2
- To = Tl = 273.15 K
- Ti = 300 K

Below is my implementation for the project. The class heatedRod can be instantiated to represent a specific rod with given properties (length, inital temp, etc.). Then, the tempMap, tempMapRK, and tempMapImplicit can be called to calculate the temperature map based upon euler explicit time stepping, RK2, or euler implicit time stepping, respectively. They utilize either cental difference finite approximation in the case of euler explicit and RK, or built-in linear algebra solver for euler implicit. Unfortunately, my implementation of the exact (to machine precision) solution is unfinished and as such, only differences between predictions will be discussed.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from scipy.integrate import quad
from scipy.sparse import diags

class heatedRod():
    
    def __init__(self, l, initT, rho, c, a, k):
        self.l = l
        self.Ti = initT
        self.rho = rho
        self.c = c
        self.a = a
        self.k = k
        self.alpha = k/(rho*c)
        
    def tempMap(self, n, dt, nt):
        self.n = n
        self.nt = nt
        self.time = (int)(dt*nt)
        print(self.time)
        dx = self.l/n
        E = np.zeros((nt, 1))
        T = np.zeros((n, nt))
        T[1:n-1 ,0] = self.Ti
        T[0,:] = 273.15
        T[n-1,:] = 273.15

        #loop over time 
        for t in range(0, nt):
            #loop over space (length)
            for i in range(1, n):
                
                if i < n-1 and t < nt-1:
                    # calc second deriv using 3pt central diff, euler explicit method
                    T[i, t+1] = T[i, t] + (self.alpha*dt/(dx**2))*(T[i-1, t] - 2*T[i, t] + T[i+1, t])
                
                # calc E from trapezoidal rule
                E[t] += (T[i-1, t] + T[i, t])*dx/2

                
        # mult internal energy inegration by rho,c,a to get actual internal energy
        E = E * (self.rho * self.c * self.a)
        
        # return the temperature matrix and the energy vector
        return T, E
    
    def tempMapRK(self, n, dt, nt):
        self.n = n
        self.nt = nt
        self.time = (int)(dt*nt)
        print(self.time)
        dx = self.l/n
        E = np.zeros((nt, 1))
        T = np.zeros((n, nt))
        T[1:n-1 ,0] = self.Ti
        T[0,:] = 273.15
        T[n-1,:] = 273.15

        #loop over time 
        for t in range(0, nt):
            #loop over space (length)
            for i in range(1, n):
                
                if i < n-1 and t < nt-1:
                    # calc second deriv using 3pt central diff, euler mod RK method
                    k1 = (self.alpha*dt/(dx**2))*(T[i-1, t] - 2*T[i, t] + T[i+1, t])
                    k2 = (self.alpha*dt/(dx**2))*(T[i-1, t+(int)(dt*k1)] - 2*T[i, t+(int)(dt*k1)] + T[i+1, t+(int)(dt*k1)])
                    T[i, t+1] = T[i, t] + (k1 + k2)/2
                
                # calc E from trapezoidal rule
                E[t] += (T[i-1, t] + T[i, t])*dx/2

                
        # mult internal energy inegration by rho,c,a to get actual internal energy
        E = E * (self.rho * self.c * self.a)
        
        # return the temperature matrix and the energy vector
        return T, E
    
    def tempMapImplicit(self, n, dt, nt):
        self.n = n
        self.nt = nt
        self.time = (int)(dt*nt)
        print(self.time)
        dx = self.l/n
        E = np.zeros((nt, 1))
        T = np.zeros((n, nt))
        T[1:n-1 ,0] = self.Ti
        T[0,:] = 273.15
        T[n-1,:] = 273.15
        
        F = self.alpha*dt/dx**2  # fourier number
        
        # Data structures for the linear system
        A = np.zeros((n, n))
        b = np.zeros(n)

        for i in range(1, n-1):
            A[i,i-1] = -F
            A[i,i+1] = -F
            A[i,i] = 1 + 2*F
        A[0,0] = A[n-1,n-1] = 1
        
        print(A.shape)
        
        #loop over time 
        for t in range(0, nt):
            if t<self.time-1:
                #loop over space (length) to set b matrix
                for i in range(1, n-1):
                    b[i] = T[i, t]
                #reset initial conditions
                b[0] = b[n-1] = 273.15
                T[:, t+1] = np.linalg.solve(A,b)
            
            for i in range(1, n):
                # calc E from trapezoidal rule
                E[t] += (T[i-1, t] + T[i, t])*dx/2
                
        # mult internal energy inegration by rho,c,a to get actual internal energy
        E = E * (self.rho * self.c * self.a)
        
        # return the temperature matrix and the energy vector
        return T, E
    
    def surfPlot(self, T):
        fig = plt.figure()
        ax = fig.gca(projection='3d')

        # Data
        X = np.arange(0, self.nt, 1)
        Y = np.arange(0, self.n, 1)
        X, Y = np.meshgrid(X, Y)

        surf = ax.plot_surface(X, Y, T, cmap=cm.coolwarm,
                               linewidth=0, antialiased=False)

        # z axis
        # ax.set_zlim(-1.01, 1.01)
        # ax.zaxis.set_major_locator(LinearLocator(10))
        # ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

        # Add a color bar which maps values to colors.
        fig.colorbar(surf, shrink=0.5, aspect=5)

        plt.show()
    
    #unused - could not get working
    def _integrand(self, _n, x):
        return (300 - 273.15)*np.sin(_n*np.pi*x/self.l)
    
    #unused - could not get analytical sol'n working
    def analytical(self, n, dt, nt, integrand):
        """ n  - number of points on bar
            dt - distance of time step
            nt - number of time steps"""
        time = (int)(dt*nt)
        T = np.zeros((n, time))
        B = np.zeros((100, 0))
        for i in B:
            B[i] = (2/self.l)*quad(integrand(i, x), 0, self.l)
        for x, t in T:
            T[x, t] = 273.15 + sum(B)
            
        return T    
        
    
```

## Spatial Resolution Results

Above, it is apparent that the increases in resolution will increase the accuracy as seen in figure 1->2. Although a stack overflow is encountered in the final scenario, and as such the results are invalid. Thus, increasing the resolution is effective as long as overflow is not encountered.

# Conclusion

- The time resolution actually has little bearing on the success of the algorithm qualitatively, although it can be assummed that increasing the time frequency will increase the accuracy. All time scales attempted as specified in the assignment produced excellent results.
- The spatical resolution will increase the accuracy as long as it maintains under overflow conditions. I encountered overflow at n=101. 

I believe the truncation in my terms in the RK implementation are causeing the similarity in results at the `K1h` term is essentially removed, making the `(k1 + k2)/2` term become simply the same as the Euler Explicit implementation. Furthermore, the error in the Euler Explicit alrogithm is quantified as follows:

- The local truncation error is on the order of h^2
- The global truncation error is on the order of h

For our examples, the error can be describes as follows:

- E(dt = .1) <= .11
- E(dt = .025) <= .025625
- E(dt = .01) <= .0101

This is reasonable, as there was little error observed in the time resolution.
